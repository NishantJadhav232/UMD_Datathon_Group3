```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Image_Data/metadata.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['keys'] = df['Keywords'].str.split(",")
```

```python
key_list=[]
```

```python
def count_keys(l):
    uniq_keys = set(l)
    for key in list(uniq_keys):
        key_lower = key.lower()
        key_list.append(key_lower)
```

```python
%%timeit
df['keys'].apply(count_keys)
```

```python
key_dict = dict()
```

```python
%%timeit
for key in key_list:
    if key in key_dict.keys():
        key_dict[key] += 1
    else:
        key_dict[key] = 1
```

```python
key_df = pd.DataFrame.from_dict(key_dict,orient='index').reset_index()
key_df.columns=['keys','count']
key_df.head()
```

```python
key_df.sort_values('count',ascending=False).head(30)
```
